## variables

What is a variable? 

A variable is used to store informration to be referenced and manipulated in a
computer program. This "information" can be as simple as one number, or a list of numbers (e.g. vector/matrix) or other more complex structures (like a list of other information, a dictionary, a string). We will start with just numbers for now.

What are variables for?
* Provide a way of labeling data with a descriptive name, so our
programs can be understood more clearly by the reader and ourselves.  
* It is helpful to think of variables as containers that hold
information.  Their sole purpose is to label and store data in memory. This data can then be used throughout your program. 

To create a new variable, enter the variable name in the command window, 
followed by an equal sign (=) and the value you want to assign to the
variable. We can then apply the operations we learned to those variables. Try it out in a cell below!

```python
# Examples
a = 2
b = 5
print(a + b)

electrical_bill = 60
gas_bill = 30
total_utilities = electrical_bill + gas_bill
print(total_utilities)
```

```python
#different from MATLAB, python allows us to define variables in a compact way
a, b, c = 7, 8, 9
print(a+b+c)
```

## other things that can be variables: lists
Really any python "object" can be stored in a variable. An object is any discrete thing in python that gets referred to by a single variable. Numbers are just one example out of many. We'll show one other basic type of python object (again, among countless others): the list.

### List
A list is simply an ordered list of other python objects. To define a list, we put the contents in square brackets [] with commas separating the list contents. Anything can go in the list---they don't have to be a similar kind of object.

```python
# List examples
my_list = [2, 3]
my_meta_list = [2, 3, [2, 3]] # a list containing a list! scandalous!
# can equivalently define by
my_meta_list = [2, 3, my_list]
```

```python
# To access an element of a list, use square brackets after the list with an index
# One of the big differences between Matlab and Python is that Python starts all indexing
# at 0.
print(my_list[0] + my_list[1])
```

```python
# If you try to access an element that doens't exist, python throws an error
my_list[2]
```

```python
# Fun trick: if you want the last element of a list, use "-1". Or more generally,
# you can index negatively from the end of a list by counting out negative numbers
my_long_list = [3, 4, 6, -2, 4, 9]
print(my_long_list[-1])
print(my_long_list[-2])
```

```python
#Remove or add elements
my_list_long = list(range(10,30,3)) #first lets create a long list
print(my_list_long)

my_list_long.remove(10) #remove an element from a list
print(my_list_long)

my_list_long.pop(0) #remove the first element
print(my_list_long)

my_list_long.append(3)
print(my_list_long)
```

```python
#sublists
my_lists = [[1,2],[3,4]]
print(my_lists[0])
print(my_lists[1])
```

# elementwise operations
Numpy arrays are flexible, and you can have numpy perform an operation on all the entries at once. Scalar multiplication above is one example of this, but almost any function can be applied "element-wise" to a numpy array.

```python
# Example: taking the square root of several numbers at once
my_array = np.array([4, 9, 25])
np.sqrt(my_array)
```

```python
# We can also sum the entries of an array to get a scalar
my_array.sum()
```

## plotting
You can't take square roots in basic python, and you can't plot stuff either. To access this power, you must import the highly reputed matplotlib.pyplot package

```python
import matplotlib.pyplot as plt #Do you recognize this syntax from numpy import above?
#(The following is just a magic keyword for jupyter notebook, don't worry about it)
%matplotlib notebook 
```

```python
# Plotting a vector
my_vector = np.array([2, 3, 6, -1])
plt.figure()
plt.plot(my_vector)
```

This above just plots the entries 0 - 3 of the vector, which are 2, 3, 6, and -1. This makes the X-axis quite boring--it's just the indices of the entires--while the Y-axis contains the actual information about the vector we're plotting.

```python
# Same plot as above but with dots instead of connecting lines.
plt.figure()
plt.plot(my_vector, '.')
```

```python
# Example with x-axis values specified. What if x and y didn't have the same shape?
x = np.array([-2, 0, -3, 1])
y = np.array([0, 1, 3, 4])
plt.figure()
plt.plot(x, y, '.')
plt.xlim([-5, 5])
plt.ylim([-5, 5])
```

```python
# Example: multiple plots on one figure
theta = np.arange(0, np.pi, 0.01)
x = np.cos(theta)
y = np.sin(theta)
plt.figure()
plt.plot(x)
plt.plot(y)
```
